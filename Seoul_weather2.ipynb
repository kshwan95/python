```python
import csv
import matplotlib.pyplot as plt
f = open('/content/drive/MyDrive/데이터분석/Unit05/seoul.csv',encoding='cp949')
data = csv.reader(f)
next(data)
aug = []
jan = []
for row in data :
  month = row[0].split('-')[1]
  if row[-1] != '':
    if month =='08':
      aug.append(float(row[-1]))
    if month == '01':
      jan.append(float(row[-1]))

plt.hist(aug,bins=100,color='r',label='Aug')
plt.hist(jan,bins=100,color='g',label='Jan')
plt.show( )
```

```python
import csv
import matplotlib.pyplot as plt
f = open('/content/drive/MyDrive/데이터분석/Unit05/seoul.csv',encoding='cp949')
data = csv.reader(f)
next(data)

month = [[],[],[],[],[],[],[],[],[],[],[],[]]
for row in data :
   if row[-1] != '':
     #월과 같은 번호의 인덱스에 월별 데이터 저장 
     month[int(row[0].split('-')[1])-1].append(float(row[-1]))

plt.boxplot(month)
plt.show()
```
