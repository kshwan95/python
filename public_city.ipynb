```python
import csv 
f = open('/content/drive/MyDrive/데이터분석/Unit07/age.csv',encoding='cp949')
data = csv.reader(f)
result = []
name = input('지역 입력해주세요:')
for row in data :
  if name in row[0]: #신도림이 포함된 행정구역 찾기 
    for i in row[3:]: #0세부터 100세까지 모든 연령에 대해 반복하기
      result.append(int(i.replace(',',''))) #해당 연령의 인구수 리스트에 순서대로 저장하기
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.rc('font',family='Gothic')
plt.title(name+'지역의 인구 구조')
plt.plot(result)
plt.show()
```

```python
f= open('/content/drive/MyDrive/데이터분석/Unit09/gender.csv',encoding='cp949')
data = csv.reader(f)
size = []

name = input('찾고 싶은 지역의 이름을 알려주세요 : ')
for row in data :
    if name in row[0] :
        m = 0
        f = 0
        for i in range(101) :
            m += int(row[i+3])
            f += int(row[i+106])
        break
size.append(m)
size.append(f)

import matplotlib.pyplot as plt
plt.rc('font', family ='Malgun Gothic')
color = ['crimson', 'darkcyan']
plt.axis('equal')
plt.pie(size, labels = ['남','여'], autopct ='%.1f%%', colors = color, startangle =90)
plt.title(name + ' 지역의 남녀 성별 비율')
plt.show()
```
